# Matplotlib

Most of the Matplotlib utilities lies under the pyplot submodule, and are usually imported under the plt alias:

```python
import matplotlib.pyplot as plt
import numpy as np

# Draw a line in a diagram from position (0,0) to position (6,250)

x_points = np.array([0, 6]) 
y_points = np.array([0, 250])

plt.plot(x_points, y_points)
plt.show()

```

You can plot as many points as you would like to. Just make sure you have the same number of points in both axis

```python
x_points = np.array([1, 2, 6, 8])
y_points = np.array([3, 8, 1, 10])

plt.plot(x_points, y_points)
plt.show()
```

If you do not specify the poins on the x-axis they will get defaulted to 0,1,2,3 etc depending upon the length of the y-axis. 

```python
y_points = np.array([3, 8, 1, 10, 5, 7])

plt.plot(y_points)
plt.show()

#The x-points in the example above is [0, 1, 2, 3, 4, 5], and the y-points are [0, 1, 4, 9, 16, 25].
```

## Matplotlib Markers

You can use teh keyword argument marker, to emphasize each point with a specified marker:

```python
y_points = np.array([3, 8, 1, 10])

plt.plot(y_points, marker = 'o')
plt.show()
```

Make each point a star 

```python
plt.plot(y_points, marker= '*')

#The following markers are supported in Matplotlib:

#'o'	Circle
#'*'	Star
#'.'	Point
#','	Pixel
#'x'	X
#'X'	X (filled)
#'+'	Plus
#'P'	Plus (filled)
#'s'	Square
#'D'	Diamond
#'d'	Diamond (thin)
#'p'	Pentagon
#'*'	Star
#'h'	Hexagon
#'H'	Hexagon
#'v'	Triangle Down
#'^'	Triangle Up
#'<'	Triangle Left
#'>'	Triangle Right
#'1'	Tri Down
#'2'	Tri Up
#'3'	Tri Left
#'4'	Tri Right
#'|'	Vline
#'_'	Hline
```

You can also use the shortcut string notation parameter to specify the marker. The parameter is also called fmt, and is written with this syntax

marker|line|color

```python
#example 

plt.plot(y_points, 'o:r')

#The line reference can be one of the following values

#'-'	Solid line
#':'	Dotted line
#'--'	Dashed line
#'-.'	Dashed/dotted line
#Therefore 'o:r' means o-markers with a dotted line, and color red

#The short color values can be one of the following

#'r'	Red
#'g'	Green
#'b'	Blue
#'c'	Cyan
#'m'	Magenta
#'y'	Yellow
#'k'	Black
#'w'	White
#The default value is 'b' (blue)
```

You can use the keyword argument markersize or ms to set the size of the markers.

```python
plt.plot(y_points, marker='o', ms=20)
```

markeredgecolor mec, and markerfacecolor mfc can be used to set the marker colors

```python
plt.plot(y_points, marker='o', ms=20, mec='b', mfc='r')
#ms = marker size
#mec = marker edge color
#mfc = marker face color
```

## Matplotlib lines

Use the keyword argument linestyle or ls, to change the style of the plotted line

```python
plt.plot(y_points, ls='dotted')

#The linestyle can be one of the following values

# 'dotted' or ':'
# 'dashed' or '--'
# 'dashdot' or '-.'
# 'solid' or '-'
# 'None' or ' ' or ''
```

You and use the keyword argument color or c to set the color of the line

```python
plt.plot(y_points, c='g')
plt.show()
```

You can use the keyword linewidth of lw to change the width of the line, The value is a floating point number

```python
plt.plot(y_points, lw=20.5)
```

## Matplotlib Lables and Titles

With pyplot you can use the xlabel() and the ylabel() functions to set a label for the x and y axis

```python
x = np.array([80, 85, 90, 95, 100, 105, 110, 115, 120, 125])
y = np.array([240, 250, 260, 270, 280, 290, 300, 310, 320, 330])

plt.plot(x, y)

plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.show()
```

Use the title() function to add a title for the plot

```python
plt.plot(x, y)

plt.title('Sports Watch Data')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.grid()

plt.show()
```

You can use the fontdict parameter in labels and titles to set the folt properties for these

```python
font1 = {'family': 'serif', 'color': 'blue' , 'size':20}
font2 = {'family': 'serif', 'color': 'darkred' , 'size':15}

plt.title('Sports Watch Data', fontdict = font1)
plt.xlabel('Average Pulse', fontdict = font2)
plt.ylabel('Calorie Burnage', fontdict = font2)

plt.grid(color = 'c', linestyle = '--', linewidth = 1)

plt.plot(x, y)

plt.show()
```

You can use the loc parameter in title() to position the title

```python
plt.title('Sports Watch Data', loc = 'left')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.grid()

plt.plot(x, y)

plt.show()
```

## Matplotlib grid

We use the grid() to add gridlines as shown above. You could also specify which gridlines to show

```python
plt.title('Sports Watch Data', loc = 'right')
plt.xlabel('Average Pulse')
plt.ylabel('Calorie Burnage')

plt.grid(axis='y')

plt.plot(x, y)

plt.show()
```
